```python
import networkx as nx
import random
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt

def genlaplacian(G, x, nodelist=None):
    if x==0:
        D = nx.laplacian_matrix(G).toarray()
    else:
        D = nx.floyd_warshall_numpy(G, nodelist)
        rowsize = D.shape[0]
        for i in range(0, rowsize):
            for j in range(i+1, rowsize):
                if(D[i,j] ==1):
                    D[i,j]= D[j,i] = -1
                else:
                    D[i,j] = D[j,i] = -D[i,j]*(pow(x, (D[i,j]-1)))
            D[i,i] = -1*(D[i].sum())
    return D

G = nx.Graph()
edgelst = [(1,3),(1,2),(1,4),(1,5),(1,6),(1,7),(1,8),(1,9),(1,10)]
G.add_edges_from(edgelst)

nodes = G.nodes()
#a = (nx.laplacian_matrix(G)).todense()
x_vals = np.arange(0,0.5,0.005)
phi_0 =[8,0,3,0,0,5,0,0,0,0]

lst = [phi_0]
t_vals =[]
for x in x_vals:
    diff = np.var(np.array(phi_0))
    a =  genlaplacian(G,x)
    t=0.05
    while (diff > 0.00001) :                 #Loop through times and decay each initial component
        phi_t = np.dot(phi_0, expm(-a*t*1))
        lst.append(phi_t)             
        diff = np.var(np.array(phi_t))
        t = t+0.05
    t_vals.append(t-0.05)
    
plt.plot(x_vals,t_vals)
plt.xlabel('x values')
plt.ylabel('Equilibrium time')
#plt.ylim(0.0,1.0)
plt.title('Diffusion of heat over a 4-regular network')
plt.legend()
plt.show()
```

```python
import numpy as np
import networkx as nx
from scipy import linalg as LA
import random

n= 100
G = nx.barabasi_albert_graph(n, 3, seed=1)

x_vals = [0,0.1,0.2,0.3,0.4]

phi_0= np.random.randint(0, 10, size=n)
phi_lst=[phi_0]  
diff = np.var(np.array(phi_0))
for x in x_vals:
    L = genlaplacian(G, x)
    D, V = LA.eigh(L);                              #Compute the eigenvalues/vectors of the laplacian matrix
    D = np.reshape(D,(n,1), order='F')
    phi_0 = np.reshape(phi_0,(n,1), order='F');
    phi_0V = np.dot((V.conj().transpose()),phi_0) #Transform the initial condition into the coordinate system of the eigenvectors

    #plotting results
    t=0.05
    while (diff > 0.0001) :                             #Loop through times and decay each initial component
        Phi = (phi_0V* (np.exp(-D*t)));                             #Exponential decay for each component
        Phi = np.dot(V,Phi);  
        Phi = (Phi.ravel()).tolist()               #shape phi to a lattice shape
        phi_lst.append(Phi)
        diff = np.var(np.array(Phi))
        t = t+0.05
    t_vals.append(t-0.05)
    
plt.plot(x_vals,t_vals)
plt.xlabel('x values')
plt.ylabel('Equilibrium time')
#plt.ylim(0.0,1.0)
plt.title('Diffusion of heat over a Barabasi(1000,3) network')
plt.legend()
plt.show()
```
