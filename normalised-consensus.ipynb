```python
import networkx as nx
import random
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt

L = np.matrix([[1,-1,0,0],[-1,3,-1,-1],[0,-1,1,0],[0,-1,0,1]])
D = np.diag([1,3,1,1])
I = np.identity(4)
L_e = I -1/9*L
x_0 = np.matrix([[1],[2],[0],[0]])
#x_0 =[1,2,0,0]


lst = []
lst.append((x_0.transpose()).tolist()[0])
time = np.arange(1,50,1)
for t in time:
    #phi_t = np.dot(phi_0, expm(-a*t))
    x_t = np.dot(L_e,x_0)
    lst.append((x_t.transpose()).tolist()[0])
    x_0 = x_t
    
print(lst[40])

plt.plot(np.arange(0,50,1),lst)
#plt.xlabel('Time, tt')
#plt.ylabel('Quantity of heat at time t, $\x_i(t)$')
plt.title('Diffusion of heat over a simple network')
plt.legend()
plt.show()
```

```python
import networkx as nx
import random
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt
from numpy.linalg import inv

L = np.matrix([[1,-1,0,0],[-1,3,-1,-1],[0,-1,1,0],[0,-1,0,1]])
D = np.diag([1,3,1,1])
I = np.identity(4)
Dv = inv(D)
L_e = I -1/9*(np.dot(Dv,L))
x_0 = np.matrix([[1],[2],[0],[0]])
#x_0 =[1,2,0,0]


lst = []
lst.append((x_0.transpose()).tolist()[0])
time = np.arange(1,50,1)
for t in time:
    #phi_t = np.dot(phi_0, expm(-a*t))
    x_t = np.dot(L_e,x_0)
    lst.append((x_t.transpose()).tolist()[0])
    x_0 = x_t

print(lst[49])
plt.plot(np.arange(0,50,1),lst)
#plt.xlabel('Time, tt')
#plt.ylabel('Quantity of heat at time t, $\x_i(t)$')
plt.title('Diffusion of heat over a simple network')
plt.legend()
plt.show()
```
