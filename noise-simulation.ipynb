```python
import networkx as nx
import numpy
import random
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt

G = nx.Graph()
edgelst = [(1,2),(1,3),(1,4),(1,5),(1,6),(1,7),(1,8),(1,9),(1,10)]
G.add_edges_from(edgelst)

nodes = G.nodes()
a = (nx.laplacian_matrix(G)).todense()

phi_0 =[8,0,3,0,0,5,0,0,0,0]
lst = []
lstn = []
time = np.arange(0,10,0.05)
mean = 0
std = 0.1 
num_samples = 10

for t in time:
    phi_t = np.dot(phi_0, expm(-a*t))
    lst.append(phi_t)
    samples = numpy.random.normal(mean, std, size=num_samples)
    lstn.append([sum(x) for x in zip(phi_t, samples)])
    
plt.plot(time,lst)
plt.xlabel('Time, t')
plt.ylabel('Quantity of heat at time t, $\phi_i(t)$')
plt.title('Diffusion of heat over a simple network')
plt.legend()
plt.show()

plt.plot(time,lstn)
plt.xlabel('Time, t')
plt.ylabel('Quantity of heat at time t, $\phi_i(t)$')
plt.title('Diffusion of heat over a simple network')
plt.legend()
plt.show()
```
